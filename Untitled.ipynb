```python
%matplotlib notebook
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
```

```python

int_slider = widgets.IntSlider(
    value=5, 
    min=0, max=10, step=1, 
    description='slider'
)
 
float_slider = widgets.FloatSlider(
    value=5, 
    min=0, max=10, step=0.01, 
    description='slider'
)
    
int_range_slider = widgets.IntRangeSlider(
    value=(20, 40), 
    min=0, max=100, step=2, 
    description='range slider'
)
 
dropdown = widgets.Dropdown(
    value='feb', 
    options=['jan', 'feb', 'mar', 'apr'], 
    description='dropdown'
)
 
radiobuttons = widgets.RadioButtons(
    value='feb', 
    options=['jan', 'feb', 'mar', 'apr'], 
    description='radio buttons'
)
 
combobox = widgets.Combobox(
    placeholder='start typing... (e.g. L or o)',
    options=['Amsterdam', 'Athens', 'Lisbon', 'London', 'Ljubljana'], 
    description='combo box'
)
 
checkbox = widgets.Checkbox(
    description='checkbox',
    value=True
)
 
 
def bla(x,y,w):
    bb=np.zeros((100,100))
    for i in range(len(x)):
        bb[:,i]=np.sin(w*x[i])+np.sin(w*y)
    return(bb)    

x = np.linspace(0, 2 * np.pi, 100)
y = np.linspace(0, 2 * np.pi, 100)

fig, ax = plt.subplots()
pp = ax.imshow(bla(x,y,1))
ax.grid(True)
 
def update(change):
    ax.clear()
    #ax.plot(x, np.sin(change.new*x))
    ax.imshow(bla(x,y,change.new))
    #line.set_ydata(np.sin(change.new * x))
    ax.grid(True)
    fig.canvas.draw()
     

float_slider.observe(update, 'value')
#int_slider.observe(update, 'value')

# a VBox container to pack widgets vertically
widgets.VBox(
    [
        float_slider, 
        int_range_slider, 
#        dropdown, 
#       radiobuttons,
#       checkbox,
#       combobox,
   ]
)
```
